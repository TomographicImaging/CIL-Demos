```python

 
import numpy as np
import matplotlib.pyplot as plt
import os
from cil.framework import Sampler
```

```python
sampler=Sampler.sequential(10)
sampler.show_epochs(5)
for _ in range(100):
    print(sampler.next())

next(sampler)
```

```python
sampler=Sampler.randomWithoutReplacement(11)
sampler.show_epochs(5)
for _ in range(30):
    print(sampler.next())
sampler.show_epochs(5)
```

```python
sampler=Sampler.hermanMeyer(12)
sampler.show_epochs(5)
for _ in range(100):
    print(next(sampler))
```

```python
sampler=Sampler.randomWithReplacement(5)
#sampler.show_epochs(5)
#for _ in range(100):
 #   print(sampler.next())
print(sampler.get_epochs())
```

```python
sampler=Sampler.randomWithReplacement(4, [0.7,0.1,0.1,0.1])
sampler.show_epochs(5)
```

```python
sampler=Sampler.staggered(20,4)
sampler.show_epochs(5)
for _ in range(100):
    print(sampler.next())
sampler.show_epochs(5)
```

```python
sampler=Sampler.customOrder([1,4,6,7,8,9,11])
sampler.show_epochs(5)
for _ in range(100):
    print(sampler.next())
sampler.show_epochs(5)
```
