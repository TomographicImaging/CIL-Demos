### Load and visualise data with NikonDataReader

This example shows how to use the `NikonDataReader` to extract data and geometry information from a Nikon .xtekct files

Here we use the dataset `korn.zip` from https://zenodo.org/record/6874123#.Y0ghJUzMKUm

Load data into an `AcquisitionData` object using the NikonDataReader

```python
from cil.io import NikonDataReader
filename = 'Korn i kasse/47209 testscan korn01_recon.xtekct'
data = NikonDataReader(file_name=filename).read()
```

View the geometry with `show_geometry` to display information about the source and detector setup

```python
from cil.utilities.display import show_geometry
show_geometry(data.geometry)
```

View the data with `show2D`

```python
from cil.utilities.display import show2D
show2D(data)
```
