### Load and visualise data with ZeissDataReader

This example shows how to use the `ZeissDataReader` to extract data and geometry information from a Zeiss .txrm files

Here we use the dataset `walnut.zip` from https://zenodo.org/record/4822516

Load data into an `AcquisitionData` object using the NikonDataReader

```python
from cil.io import ZEISSDataReader
filename = "valnut/valnut_2014-03-21_643_28/tomo-A/valnut_tomo-A.txrm"
data = ZEISSDataReader(file_name=filename).read()
```

View the geometry with `show_geometry` to display information about the source and detector setup


```python
from cil.utilities.display import show_geometry
show_geometry(data.geometry)
```

View the data with `show2D`

```python
from cil.utilities.display import show2D
show2D(data)
```
